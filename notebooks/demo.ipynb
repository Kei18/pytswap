```python
%load_ext autoreload
%autoreload 2

from pytswap import (
    TSWAP,
    get_grid,
    get_scenario,
    is_valid_amapf_solution,
    save_configs_for_visualizer
)
```

```python
map_file = "../assets/random-32-32-10.map"
scen_file = "../assets/random-32-32-10-random-1.scen"
N = 200

grid = get_grid(map_file)
starts, targets = get_scenario(scen_file, N)
```

```python
%%time
planner = TSWAP(grid, starts, targets)
solution = planner.solve()
```

```python
is_valid_amapf_solution(grid, starts, targets, solution)
```

```python
result_file = "tmp.txt"
save_configs_for_visualizer(solution, result_file)
```
